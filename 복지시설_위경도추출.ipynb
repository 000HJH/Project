```python
#주소 -> 위경도 추출, 오류나지 않고 넘어감
from geopy.geocoders import Nominatim

def geocoding(address):
    geolocoder = Nominatim(user_agent = 'South Korea', timeout=None)
    geo = geolocoder.geocode(address)
    crd = {"lat": str(geo.latitude), "lng": str(geo.longitude)}

    return crd
```

```python
# csv에 쓰기 

df = pd.read_csv('./data/my복지관.csv')
adds = df.iloc[:,1] #도로명주소 컬럼
p = r'(\D[^(,^,]+)'

lat_lst=[]
lng_lst=[]
for idx,add in enumerate(adds):
    new_add= re.findall(p,add)[0] #주소 정리
    try:
        crd = geocoding(new_add) #주소로 위경도 탐색
        lat_lst.append(crd['lat']) #위도
        lng_lst.append(crd['lng']) #경도

    except:                 #(오류나면)못찾으면 공백 입력
        lat_lst.append(' ') #위도
        lng_lst.append(' ') #경도

df['lat'] = lat_lst
df['lon'] = lng_lst

df.to_csv('./data/my복지관_xy.csv', encoding='cp949') #새 csvㄹ
```
